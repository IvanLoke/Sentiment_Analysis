```python
from bs4 import BeautifulSoup
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
list_of_names = []
list_of_text = []
for i in range(1 ,37):
  if i == 1:
    i = ""
  f = open(f"/content/drive/MyDrive/Work 2022/REP_Analytics/ChatExport_2022-05-10/messages{i}.html")
  contents = f.read()
  soup = BeautifulSoup(contents,'lxml')
  user = ""
  for big_chunk in soup.find_all('div', {"class": "message"}):
    if "service" not in big_chunk["class"]:
      book = big_chunk.find('div',{"class":"text"})
      if book == None:
        continue
      book = book.get_text()
      book = book.replace("\n","").strip()
      list_of_text.append(book)
      google = big_chunk.find('div',{"class":"from_name"})
      if google != None:
        user = google.get_text()
      user = user.replace("\n","").strip()
      list_of_names.append(user)
print(len(list_of_names))
print(len(list_of_text))

  #   if google == None:
  #     user = user.replace("\n","").strip()
  #     list_of_names.append(user)
  #     print(user)
  #   else:
  #     user = google.get_text()
  #     user = user.replace("\n","").strip()
  #     list_of_names.append(user)
  #     print(user)
  # print(list_of_names)
    # for div in big_chunk.find_all('div',{"class":"from_name"}):
    #   divstring = div.get_text()
    #   print(divstring)

```
